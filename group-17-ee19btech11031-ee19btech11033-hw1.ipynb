# Question 1(b)

```python

import matplotlib.pyplot as plt
import numpy as np
import wave


print("Enter the name of the original .wav file(Along with the extension):")
str_1=input()
inp = wave.open(str_1,mode='rb')
data = inp.readframes(-1)
r = inp.getframerate()#Sampling rate
inp.close()
r1 = 2*r
r2 = r
r3 = 0.5*r
r4 = 0.25*r
t = 0.02 #20 milliseconds
n = int(np.floor(r*t))
n1 = int(np.floor(r1*t))
n2 = int(np.floor(r2*t))
n3 = int(np.floor(r3*t))
n4 = int(np.floor(r4*t))

def DTFT(y,M,a,N,W):
    Y = [0 for i in range(N)]
    for i in range(N):
        for k in range(M):
            Y[i] += y[k]*np.exp((-1j)*W[i]*(a+k))
    return Y

y = [0 for i in range(n)]
for i in range(n):
    y[i] = data[i]

inp1 = wave.open(str_1[:-4]+"_1.wav",mode='rb')#file of the form wow_1.wav(sampling frequency=2*r)
data1 = inp1.readframes(-1)
inp1.close()
y1 = [0 for i in range(n1)]
for i in range(n1):
    y1[i] = data1[i]

inp2 = wave.open(str_1[:-4]+"_2.wav",mode='rb')#file of the form wow_2.wav(sampling frequency=r)
data2 = inp2.readframes(-1)
inp2.close()
y2 = [0 for i in range(n2)]
for i in range(n2):
    y2[i] = data2[i]

inp3 = wave.open(str_1[:-4]+"_3.wav",mode='rb')#file of the form wow_3.wav(sampling frequency=0.5*r)
data3 = inp3.readframes(-1)
inp3.close()
y3 = [0 for i in range(n3)]
for i in range(n3):
    y3[i] = data3[i]

inp4 = wave.open(str_1[:-4]+"_4.wav",mode='rb')#file of the form wow_4.wav(sampling frequency=0.25*r)
data4 = inp4.readframes(-1)
inp4.close()
y4 = [0 for i in range(n4)]
for i in range(n4):
    y4[i] = data4[i]

a = 1
N = 1000
W = np.linspace((-1)*np.pi, np.pi, num=N, endpoint=False)

Y = DTFT(y,n,a,N,W)#original frequency:r
Y_mag = [0 for i in range(N)]
for i in range(N):
    Y_mag[i] = np.abs(Y[i]) #Magnitude
Y_ph = [0 for i in range(N)]
for i in range(N):
    Y_ph[i] = np.angle(Y[i], deg=True) #Phase

Y1 = DTFT(y1,n1,a,N,W)#2*r
Y_mag1 = [0 for i in range(N)]
for i in range(N):
    Y_mag1[i] = np.abs(Y1[i])
Y_ph1 = [0 for i in range(N)]
for i in range(N):
    Y_ph1[i] = np.angle(Y1[i], deg=True)

Y2 = DTFT(y2,n2,a,N,W)#r
Y_mag2 = [0 for i in range(N)]
for i in range(N):
    Y_mag2[i] = np.abs(Y2[i])
Y_ph2 = [0 for i in range(N)]
for i in range(N):
    Y_ph2[i] = np.angle(Y2[i], deg=True)

Y3 = DTFT(y3,n3,a,N,W)#r*0.5
Y_mag3 = [0 for i in range(N)]
for i in range(N):
    Y_mag3[i] = np.abs(Y3[i])
Y_ph3 = [0 for i in range(N)]
for i in range(N):
    Y_ph3[i] = np.angle(Y3[i], deg=True)

Y4 = DTFT(y4,n4,a,N,W)#r*0.25
Y_mag4 = [0 for i in range(N)]
for i in range(N):
    Y_mag4[i] = np.abs(Y4[i])
Y_ph4 = [0 for i in range(N)]
for i in range(N):
    Y_ph4[i] = np.angle(Y4[i], deg=True)

plt.subplot(2,1,1)
plt.stem(W,Y_mag,label=str_1+'(Magnitude)')
plt.xlabel('w')
plt.legend()
plt.subplot(2,1,2)
plt.stem(W,Y_ph,label=str_1+'(Phase)')
plt.xlabel('w')
plt.legend()
plt.show()

plt.subplot(2,1,1)
plt.stem(W,Y_mag1,label=str_1[:-4]+'_1.wav'+'(Magnitude)')
plt.xlabel('w')
plt.legend()
plt.subplot(2,1,2)
plt.stem(W,Y_ph1,label=str_1[:-4]+'_1.wav'+'(Phase)')
plt.xlabel('w')
plt.legend()
plt.show()

plt.subplot(2,1,1)
plt.stem(W,Y_mag2,label=str_1[:-4]+'_2.wav'+'(Magnitude)')
plt.xlabel('w')
plt.legend()
plt.subplot(2,1,2)
plt.stem(W,Y_ph2,label=str_1[:-4]+'_2.wav'+'(Phase)')
plt.xlabel('w')
plt.legend()
plt.show()

plt.subplot(2,1,1)
plt.stem(W,Y_mag3,label=str_1[:-4]+'_3.wav'+'(Magnitude)')
plt.xlabel('w')
plt.legend()
plt.subplot(2,1,2)
plt.stem(W,Y_ph3,label=str_1[:-4]+'_3.wav'+'(Phase)')
plt.xlabel('w')
plt.legend()
plt.show()

plt.subplot(2,1,1)
plt.stem(W,Y_mag4,label=str_1[:-4]+'_4.wav'+'(Magnitude)')
plt.xlabel('w')
plt.legend()
plt.subplot(2,1,2)
plt.stem(W,Y_ph4,label=str_1[:-4]+'_4.wav'+'(Phase)')
plt.xlabel('w')
plt.legend()
plt.show()

print("From the graphs, we can see that as the sampling frequency decreases, the magnitude of the low frequency components relative to the higher frequency components increases. This explains why the pitch appears to become lower. We can also see that the magnitude of the frequency domain components decreases as the sampling frequency decreases.")
```

# Question 1(c)

```python
import numpy as np
import matplotlib.pyplot as plt

x_1 = np.linspace(-2,2,41)#band limit of the signal
x = np.linspace(-2,6,2*len(x_1)-1)#w values
#frequency domian representation of the signal
y = [1 for i in range(len(x_1))]
w = [0 for i in range(len(x_1)-1)]

y_1 = y+w #fourier spectrum at omega=0 
y_2 = w+y #shifted version of y_1
z = [y_1[i]+y_2[i] for i in range(len(y_1))]#spectrum after adding both of them

plt.subplot(2,1,1)
plt.stem(x,y_1)
plt.subplot(2,1,2)
plt.stem(x,z)
plt.show()

print("Therefore, we can see that when the magnitude at the end-point of the band-limited frequency-domain representation is non-zero, then if the equality is chosen, then the original signal and the shifted signal overlap at exactly one point(i.e., the end-point), causing aliasing. Here, the frequency domain signal is taken to be of constant value 1 between -2 and 2 and 0 elsewhere.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import wave
from playsound import playsound

def sum_n(X,H,n): #convolution sum for a particular n
	su=0
	for k in range(len(X)):
		if (n-k)>=0 and (n-k)<len(H):
			su=su+X[k]*H[n-k]
	return su
		
			
	
def  convolution_1(X,H,framerate,nchannels,sampwidth,choice,str_1):
	Y=[]	
	for i in range(len(H)+len(X)-1):
		Y.append(sum_n(X,H,i))
    
	for i in range(len(Y)):
		if Y[i]<0:
			Y[i] = 0
		elif Y[i]>255:
			Y[i] = 255
		else:
			Y[i] = int(Y[i])

	Y=np.asarray(Y)	 #list to array
	Y_1=(Y).astype('b') #converting into bytes
	y_n=wave.open(str_1[:-4]+"_conv_with_"+str(choice)+"_seq"+".wav","w")#file of the form: wow_conv_with_1_seq.wav
	y_n.setnchannels(nchannels) #writing the data to a .wav file
	y_n.setsampwidth(sampwidth)
	y_n.setframerate(framerate)
	y_n.writeframes(Y_1.tobytes())
	y_n.close()
	time=np.arange(0,len(Y)/framerate,1/framerate)
	plt.xlabel("t") 
	plt.plot(time, Y) 
	plt.show()
	playsound(str_1[:-4]+"_conv_with_"+str(choice)+"_seq"+".wav") #playing the .wav file


def samp_create_play(str_1):
	raw=wave.open(str_1)
	params = raw.getparams()
	nchannels, sampwidth, framerate, nframes = params[:4]
	signal = raw.readframes(-1) #reading all frames
	signal = np.frombuffer(signal, dtype="uint8") 
	raw.close()
	time = np.arange( 0, len(signal) / framerate,1/framerate) 
	T=(time<=1).astype(np.int) #for time less than 1s
	signal_1=signal[:sum(T)] #data for which t<1s
	h_1 = [0.33,0.33,0.33]
	h_2 = [-0.25,0.50,-0.25]
	print("1.Output with first sequence:")
	print("2.Output with second sequence:")
	print("Enter choice:")
	ch=int(input())
	if ch==1:
		convolution_1(signal_1,h_1,framerate,nchannels,sampwidth,ch,str_1)
	if ch==2:
		convolution_1(signal_1,h_2,framerate,nchannels,sampwidth,ch,str_1)		

print("Enter the name of the .wav file to be sampled (Along with the extension):")
str_1=input()
samp_create_play(str_1)

print("When we convolve the wav file with the first sequence, we are adding the original signal with two time shifted signals, one by 1/f and the other by 2/f respectively. since these values are small, the blur introduced will be small. And also, since the three impulses' magnitudes add up to nearly one, we dont hear a significant decrease in the amplitude.\n\nThe second sequence has impulses with negative signs. So, when we convolve the wav signal with it, the magnitude decreases and we can observe the decrease in the amplitude. The convolution may lead to negative values but the wav file can only take values from 0 to 255. So, we need to clip the values, and this may introduce some error or noise into the signal.")
```
