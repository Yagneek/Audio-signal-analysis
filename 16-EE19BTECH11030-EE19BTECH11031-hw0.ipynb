__List of variable names :__<br>
M - Number of sample points in input signal<br>
y - values of input signal<br>
x - corresponding n values of input signal<br>
a - corresponding n value of the first element of y<br>
N - Number of values of $\omega$<br>

Question 1

```python
import numpy as np
from matplotlib import pyplot as plt

count=0
M = int(input("enter number of sample points:"))

x= [i for i in range(M)]
print("enter the samples:")
y = [complex(input()) for i in range(M)]

#Checking for complex input
for i in range(M):
    if np.imag(y[i]) != 0:
        count=count+1

if count >0:
    print('signal is complex')
    
    y_mag = [0 for i in range(M)]
    for i in range(M):
        y_mag[i] = np.abs(y[i])

    y_ph = [0 for i in range(M)]
    for i in range(M):
        y_ph[i] = np.angle(y[i], deg=True)
    
    plt.subplot(1,2,1)
    plt.stem(x,y_mag)
    plt.subplot(1,2,2)
    plt.stem(x,y_ph)
    plt.show()
    
elif count ==0:
    print('signal is real')
    
    plt.stem(x,np.real(y))
    plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

M = 101
a = -5
c = float(input("Enter the value of c : "))
d = float(input("Enter the value of d : "))

x = [i for i in np.linspace(a,-a,num=M)]
#Defining the function
y = [np.round(np.exp((c + d*1j)*i),decimals=4) for i in x]

#Checking for complex input
for i in range(M):
    if np.imag(y[i]) != 0:
        count=count+1

if count >0:
    print('signal is complex')
    
    y_mag = [0 for i in range(M)]
    for i in range(M):
        y_mag[i] = np.abs(y[i])

    y_ph = [0 for i in range(M)]
    for i in range(M):
        y_ph[i] = np.angle(y[i], deg=True)
    
    plt.subplot(1,2,1)
    plt.stem(x,y_mag)
    plt.subplot(1,2,2)
    plt.stem(x,y_ph)
    plt.show()
    
elif count ==0:
    print('signal is real')
    
    plt.stem(x,np.real(y))
    plt.show()
```

Question 3

```python
import numpy as np
import matplotlib.pyplot as plt

M = int(input("Enter the number of sample points : "))
print("Enter the samples :")
y = [complex(input()) for i in range(M)]
a = int(input("Enter the corresponding n value of the first sample : "))
N = 100
W = np.linspace((-1)*np.pi, np.pi, num=N, endpoint=False)
Y = [0 for i in range(N)]

#Calculating DTFT
for i in range(N):
    for k in range(M):
        Y[i] += y[k]*np.exp((-1j)*W[i]*(a+k))
        Y[i] = np.round(Y[i],decimals=4)

Y_mag = [0 for i in range(N)]
for i in range(N):
    Y_mag[i] = np.abs(Y[i])

Y_ph = [0 for i in range(N)]
for i in range(N):
    Y_ph[i] = np.angle(Y[i], deg=True)

plt.subplot(2,1,1)
plt.plot(W,Y_mag)
plt.subplot(2,1,2)
plt.plot(W,Y_ph)
plt.show()
```

Question 4(a)

```python
n = 10
y = [i+1 for i in range((-1)*n,n+1,1)]

M = 2*n + 1
a = (-1)*n
N = 101
W = np.linspace((-1)*np.pi, np.pi, num=N)
Y = [0 for i in range(N)]

#Calculating DTFT
for i in range(N):
    for k in range(M):
        Y[i] += y[k]*np.exp((-1j)*W[i]*(a+k))
        Y[i] = np.round(Y[i],decimals=4)

Y_conj = [0 for i in range(N)]
for i in range(N):
    Y_conj[i] = np.conjugate(Y[i])

plt.subplot(2,2,1)
plt.plot(W,np.real(Y))
plt.subplot(2,2,2)
plt.plot(W,np.imag(Y))
plt.subplot(2,2,3)
plt.plot(-W,np.real(Y_conj))
plt.subplot(2,2,4)
plt.plot(-W,np.imag(Y_conj))

#Verifying the property
equal = True
for i in range(N):
    equal = equal and np.real(Y[i]) == np.real(Y_conj[N-1-i]) and np.imag(Y[i]) == np.imag(Y_conj[N-1-i])

print("equal = ", equal)
```

Since equal = True, $X(e^{j\omega}) = X^*(e^{-j\omega})$<br>
We can also see this from the graph
